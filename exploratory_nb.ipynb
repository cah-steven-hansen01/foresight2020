```python
%%time
product_df = pd.DataFrame()
i=0
for product, frame in clean_df.groupby(by='Product'):
    product_df.loc[i,'Product'] = product
    product_df.loc[i,'number of lots'] = len(frame)
    product_df.loc[i,'number of NCs'] = sum(frame['NC'])
    product_df.loc[i,'average size of lot'] = sum(frame['requested_qty'])/len(frame['requested_qty'])
    product_df.loc[i, 'NC rate'] = sum(frame['NC'])/len(frame)
    i+=1
plt.figure(figsize = (20,5))
product_df[product_df['NC rate']>0.083]\
                    .sort_values(by = 'NC rate',ascending=False)\
                    .plot(kind='bar',x = 'Product',y = 'NC rate')
plt.title('NC Rate by Product')
plt.show()
```

```python
product_df.plot(kind ='scatter',x = 'NC rate',y = 'number of lots')
```
