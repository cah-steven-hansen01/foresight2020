## 1. archived_so_clean <a name = "archived_so_clean"></a>

```python
# create dic for components
component_category = pd.read_csv('./reference_data/Product Category.csv',encoding='windows-1252')
component_dic = {}
for component, category in zip(component_category['Component'],component_category['Category']):
    component_dic[component] = category
component_dic['R98178']
```

```python
%%time
# Pull NC data
os.chdir(r'C:\Users\steven.hansen01\Documents\GitHub\quality_productivity_tool\qa_productivity_tool')
import qa_productivity_tool.nc_full as nc_full
os.chdir(r'C:\Users\steven.hansen01\Documents\GitHub\quality_issue_predictor')
ncs = nc_full.NC_Full()
ncs.mostrecentreport()
nc_df = ncs.get_lot_numbers()
nc_df.head()
```

```python
%%time
# prepare data for one-hot encoding
# read in raw data
archived_so = pd.read_csv('./raw_data/archived_SO.csv')
archived_so.head()

not_found_components = set()
def encode(frame):
    global not_found_components
    frame = frame.reset_index(drop=True)
    lot_number = frame.loc[0,'lot_number'].strip()
    product = frame.loc[0,'Product'].strip()
    date = frame.loc[0,'Date']
    shop_order = str(frame.loc[0,'shop_order'])
    req_qty = frame.loc[0,'requested_qty']
    df = pd.DataFrame()
    df.loc[lot_number,'Product'] = product
    df.loc[lot_number,"Date"] = date
    df.loc[lot_number,'Shop Order'] = shop_order
    df.loc[lot_number,'requested_qty'] = req_qty
    for i,component in enumerate(frame.component):
        component = component.strip()
        try:
            component_dic[component]
        except KeyError:
            not_found_components.add(component)
            continue
        df.loc[lot_number,component_dic[component]] = component
    return df.reset_index()\
            .rename(columns = {'index':'Lot Number'})

list_of_frames = [encode(frame) for lot, frame in archived_so.groupby(by = 'lot_number')]
archived_so_clean = pd.concat(list_of_frames)\
            .reset_index(drop = True)
lot_mask = archived_so_clean['Lot Number'].str.contains('TST|TSA|TST')
archived_so_clean = archived_so_clean[~lot_mask]
product_mask = archived_so_clean["Product"].str.contains('RM[0-9]{3}|5551500055') # removes regrind and tips codes (see outlier investigation)
archived_so_clean = archived_so_clean[~product_mask]
# combine NC and SO data
lot_checker = lambda lot: 1*(lot.strip() in set(nc_df['Lot Number']))
archived_so_clean['NC'] = list(map(lot_checker,archived_so_clean['Lot Number']))
print('number of NCs matched = {}'.format(sum(archived_so_clean.NC)))


print('{} rows, {} columns'.format(archived_so_clean.shape[0],archived_so_clean.shape[1]))
print('{} components not found'.format(len(not_found_components)))
archived_so_clean.head()
```

```python
archived_so_clean.to_csv('./clean_data/archived_so_clean.csv')
```

## archived_so_encoded <a name="archived_so_encoded"></a>

* one-hot encode all feature columns except requeted_qty
* requested_qty is evaluated and RobustScalar normalization method is selected

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
cols = ['Shop Order','Lot Number','requested_qty','Product','packaging', 'resin', 'colorant', 'Formula Sheet', 'misc',
       'lid component lid', 'label', 'lid assembly', 'PRINTER ACCESSORIES',
       'SILICON FLUID', 'rubber tip', 'ifu', 'pad', 'label stock', 'label kit',
       'purging solvent', 'lid component hood', 'lid component', 'PAD',
       'LABEL', 'barrel', 'lid component rotor', 'lid', 'lid component door',
       'lid component cap', 'lid component washer', 'container', 'TRAY', 'fg',
       'lid component cowl', 'bracket key', 'cannula', 'tip cap']
archived_so_encoded = archived_so_clean[cols].copy()

hot_code_cols= archived_so_encoded.iloc[:,3:-1]
one_hot_encoder = OneHotEncoder(sparse=False,handle_unknown='ignore')
one_hot_encoder.fit(hot_code_cols)
df_encoded = one_hot_encoder.transform(hot_code_cols)

cols = []
i=0
while i < 33:
    cols = cols+list(one_hot_encoder.categories_[i])
    i+=1
df_encoded = pd.DataFrame(data=df_encoded, columns=cols)
df_encoded = df_encoded.loc[:, df_encoded.columns.notnull()]

archived_so_encoded = archived_so_encoded.iloc[:,0:3].join(df_encoded)
archived_so_encoded = archived_so_encoded.set_index('Shop Order')
```

#### Evaluate continous value 'requested_qty'

```python
req_qty = archived_so_encoded['requested_qty']
print('max requested_qty = {}'.format(max(req_qty)))
print('avg requested_qty = {}'.format(round(sum(req_qty)/len(req_qty)),0))
print('std requetsed_qty = {}'.format(round(np.std(req_qty),0)))
archived_so_encoded['requested_qty'].plot.hist()
plt.title('Hist Plot of requested_qty values')
plt.show()
```

##### Outlier Investigation

```python
cols = ['Lot Number','Product','Date','Shop Order','requested_qty']
archived_so_clean[cols][archived_so_clean['requested_qty']>100000].sort_values(by = 'requested_qty',ascending=False)

```

#### Scaling of requested_qty

* RobustScaler. used because it deals with outliers well and the remaining outliers will stay

```python
from sklearn.preprocessing import RobustScaler
```

```python
scaler = RobustScaler(with_centering = False).fit(archived_so_encoded['requested_qty'].to_numpy().reshape(-1,1))
archived_so_encoded['scaled_requested_qty'] = scaler.transform(archived_so_encoded['requested_qty'].to_numpy().reshape(-1,1))
archived_so_encoded.head()
```

```python
%%time
# combine NC and SO data
lot_checker = lambda lot: 1*(lot.strip() in set(nc_df['Lot Number']))
archived_so_encoded['NC'] = list(map(lot_checker,archived_so_encoded['Lot Number']))
print('number of NCs matched = {}'.format(sum(archived_so_encoded.NC)))
archived_so_encoded.head(3)
```
